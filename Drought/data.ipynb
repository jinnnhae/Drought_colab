```python
Drought = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/광주가뭄지수.csv', encoding = "utf-8", engine='python')
Drought.head()
```

```python
JAdam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/주암댐.csv', encoding = "utf-8", engine='python')
JAdam.head()
```

```python
JA = JAdam[['일시','저수율(%)', '강우량(mm)']]
juam = JA.rename(columns={'저수율(%)':'주암댐저수율(%)'})
juam.head()
```

```python
DBwater = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/DBwater.csv', encoding = "utf-8", engine='python')
date = DBwater['Datetime'].str.slice(0,10)
DBwater['일시'] = date
dongbok = DBwater.rename(columns={'저수율':'동북댐저수율(%)'})
dongbok.head()
```

```python
Rain = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/광주일강수량.csv', encoding = "utf-8", engine='python')
Rain.head()
```

```python
a = pd.merge(Drought, juam,  on='일시', how="outer") # 가뭄지수와 주암댐수치 merge
b = pd.merge(dongbok, Rain,  on='일시', how="outer") # 동복댐수치와 강우량 merge
```

```python
data = pd.merge(a, b,  on='일시', how="outer") # 모든 데이터 '일시'를 기준으로 병합
data = data[['일시', 'SPI1', 'SPI2', 'SPI3', 'SPI4', 'SPI5', 'SPI6','주암댐저수율(%)','강우량(mm)', '동북댐저수율(%)']]
```
